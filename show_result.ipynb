```python
import matplotlib.pyplot as plt
import numpy as np
import json
```

```python
with open("history_300eps.json", 'r') as file:
    data = json.load(file)
```

```python
losses_g = data['iter_losses_g']
losses_d = data['iter_losses_d']
epochs = range(1, len(losses_g) + 1)
```

```python
plt.figure(figsize=(10,5))
plt.plot(epochs, losses_g, label='G')
plt.plot(epochs, losses_d, label='D')
plt.xlabel('Iteration')
plt.ylabel('Loss')

# Add a title
plt.title('Generator and Disriminator Loss During Training')

# Add a legend to differentiate between the two losses
plt.legend()

# Show the plot
plt.savefig('G_D_losses_per_iter.png', bbox_inches='tight')
plt.show()
```

```python
from PIL import Image
import os
# List of image file paths
folder = 'process_300eps'
image_list = [os.path.join(folder, file_path) for file_path in os.listdir(folder) if file_path.endswith(".jpg") or file_path.endswith(".png")]

# Open images and store them in a list
images = [Image.open(img) for img in image_list]

# Save as GIF
images[0].save('training_process_300eps.gif',
               save_all=True,
               append_images=images[1:], 
               duration=200,  # Duration between frames (in milliseconds)
               loop=1)  

print("GIF created successfully!")
```
